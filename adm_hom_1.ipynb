```python
#Insertion Sort - Part 1

def print_to_str(l):
    string=""
    for i in range(len(l)):
        string=string+' '+str(l[i])
    return string[1:]

n=int(input())
l=list(map(int,raw_input().split()))
a=l[-1]
i=1
while a<l[n-1-i] and i<=n-1:
    l[n-i]=l[n-1-i]
    print(print_to_str(l))
    i+=1
l[n-i]=a
print(print_to_str(l))
```

```python
#Insertion Sort - Part 2

def print_to_str(l):
    string=""
    for i in range(len(l)):
        string=string+' '+str(l[i])
    return string[1:]

def insertSort2(n,arr):
    sol=""
    for i in range(1,len(arr)):
        piv=arr[i]
        j=i-1
        while j>=0 and arr[j]>piv:
            arr[j+1]=arr[j]
            j=j-1
        arr[j+1]=piv
        sol+=print_to_str(arr)+'\n'
    return sol

n=int(input())
arr=list(map(int,raw_input().split()))
sol=insertSort2(n,arr)
print(sol)
```

```python
#Alphabet Rangoli

import string

def print_rangoli(size):
    alpha = string.ascii_lowercase
    l = []
    for i in range(size):
        s = "-".join(alpha[i:size])
        l.append((s[::-1]+s[1:]).center(4*size-3, "-"))
    print(('\n'.join(l[:0:-1]+l)))
```
